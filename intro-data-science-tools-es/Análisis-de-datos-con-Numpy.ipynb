#Numpy

El elemento básico en Numpy es el array, o mejor dicho, el **ndarray**. La diferencia de estos arrays con los de Python reside en los siguientes puntos:

* Siven como extensión a los arrays tradicionales de Python para dar un mejor soporte a los arrays multidimensionales y acercarlos a un mejor procesamiento a nivel de hardware.
*  En general, se comportan y disponen de primitivas (operadores y constructores) para la computación eficiente de arrays (habitual en computación científica)

* [Sintaxis oficial para slicing](https://docs.scipy.org/doc/numpy-1.13.0/reference/arrays.indexing.html).

The basic slice syntax is i:j:k where i is the starting index, j is the stopping index, and k is the step (k\neq0). This selects the m elements (in the corresponding dimension) with index values i, i + k, ..., i + (m - 1) k where m = q + (r\neq0) and q and r are the quotient and remainder obtained by dividing j - i by k: j - i = q k + r, so that i + (m - 1) k < j.

```
#Los n primeros elementos.
print(a[:5])
#Los len-n últimos elementos.
print(a[5:])
```

```
#Los elementos entre k:m, [k,m)
a[4:6]
```

```
#Seleccionando filas y columnas
a = np.diag(np.arange(5))
a[2,1] = 100
print(a)
k = 2

#Fila i = 2
print ("Fila "+str(k))
print (a[k])
print ("Columna "+str(k))

#Columna j = 2
print (a[:,k])

#N primeras filas
print ("N Filas: "+str(k))
print (a[:k])

#N primeras filas
print ("N primeras filas: "+str(k))
print (a[:k,])

#Selección desde la fila N en adelante
print ("Selección de filas: "+str(k))
print (a[k:])
      
#Otras formas
#Columna k
print (a[...,k])
#Fila k
print (a[k,...])
```

```
#Comparaciones entre arrays.

a = np.arange(5)
b = np.arange(5)
print (a == b)
print (a > b)

a = np.array([1, 1, 0, 0], dtype=bool)
b = np.array([1, 0, 1, 0], dtype=bool)
print(np.logical_or(a, b))
print (np.logical_and(a, b))

```

```
#Transpuesta

x = np.arange(4).reshape((2,2))
print(x)
#¡¡¡Es sólo una vista!!!
print(np.transpose(x))
print(x.T)
```

```
#Agregación de valores:

x = np.array([1, 2, 3, 4])
print(str(np.sum(x)))

x = np.array([[1, 2], [3, 4]])
print (x)

#Sumar por columnas
print (str(x.sum(axis=0)))
print (str(x[:, 0].sum())+"  "+str(x[:, 1].sum()))

#Sumar por filas
print (str(x.sum(axis=1)))
print (str(x[0, :].sum())+"  "+str(x[1, :].sum()))
```

* **Sparse matrices**

```
import numpy as np
from scipy.sparse.linalg import eigsh
from scipy.linalg import eigh
import scipy.sparse
import time

N = 3000

# Se crear una matriz sparse
m = scipy.sparse.rand(N, N)

# Se clona
a = m.toarray()

print('Tamaño: ' + str(a.nbytes) + ' bytes')
print('Tamaño como sparse: ' + str(m.data.nbytes) + ' bytes')

# Eficiencia
t0 = time.time()

res1 = eigh(a)
dt = str(np.round(time.time() - t0, 3)) + ' seconds'
print('Non-sparse: ' + dt)

# Sparse
t0 = time.time()
res2 = eigsh(m)
dt = str(np.round(time.time() - t0, 3)) + ' seconds'
print('Sparse: ' + dt)
```

* **Estadística descriptiva**

```
import numpy as np

# Array de 1000 elementos aleatorios con una distribución normal.
x = np.random.randn(1000)

# Algunas estadísticas descriptivas
mean = x.mean()
std = x.std()
var = x.var()
```

```
import numpy as np
from scipy import stats

# Array de 100 elementos aleatorios con una distribución normal.
sample = np.random.randn(100)

# Media armónica
out = stats.hmean(sample[sample > 0])
print('Hmena = ' + str(out))

# Media entre valores 
out = stats.tmean(sample, limits=(-1, 1))
print('\nTmean = ' + str(out))



# Describiendo...
out = stats.describe(sample)
print('\nSize = ' + str(out[0]))
print('Min = ' + str(out[1][0]))
print('Max = ' + str(out[1][1]))
print('Mean = ' + str(out[2]))
print('Var = ' + str(out[3]))
print('Skewness = ' + str(out[4]))
print('Kurtosis = ' + str(out[5]))
```

* **Distribuciones estadísticas**

```
import subprocess
import numpy as np
import scipy.stats as s
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(8, 9))
fig.subplots_adjust(hspace=0, wspace=0)

# Distribuciones
dists = ['norm', 'lognorm', 'gamma', 'invgauss', 'cauchy',
         'logistic', 'maxwell', 'powerlaw', 'rdist',
         'wald', 'alpha', 'rayleigh', 'triang', 'lomax',
         'laplace', 'gilbrat', 'fisk', 'erlang', 't', 'nakagami']

dists = np.sort(dists)

for i, D in enumerate(dists):
    x = np.linspace(-5, 5, 1000)
    func = s.__dict__[D]

    try:
        dist = func(loc=0)
        pdf = dist.pdf(x)
    except:
        dist = func(0.7)
        pdf = dist.pdf(x)

    ax = fig.add_subplot(5, 4, i + 1)
    D = (D[0]) + D[1::]
    ax.plot(x, pdf, label=D)
    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False)
    ax.set_xlim(-4, 4)
    ax.set_ylim(0, pdf.max() * 1.5)
    ax.legend(loc='upper right', markerscale=0.0001, prop={'size': 10})

    leg = plt.gca().get_legend()
    ltext = leg.get_texts()
    llines = leg.get_lines()
    frame = leg.get_frame()
    plt.setp(llines, linewidth=0)
    leg.draw_frame(False)
```

* **Interpolación**

```
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

# Creaación de datos.
x = np.linspace(0, 10 * np.pi, 20)
y = np.cos(x)

# Tipos de interopolación
fl = interp1d(x, y, kind='linear')
fq = interp1d(x, y, kind='quadratic')

# x.min and x.max para no salirse de los límites
xint = np.linspace(x.min(), x.max(), 1000)
yintl = fl(xint)
yintq = fq(xint)


fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(111)
ax.plot(xint, yintl, label='Linear')
ax.plot(xint, yintq, label='Quadratic')
ax.scatter(x, y, marker='o',
           edgecolor='red', facecolor='none', s=50)
ax.legend(loc='upper left')

ax.set_xlim(0, 10 * np.pi)
ax.set_ylim(-2, 2)
```

* **Optimización**

```
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Función para crear los datos
def func(x, a, b):
    return a * x + b

# Greneración de los datos
x = np.linspace(0, 10, 100)
y = func(x, 1, 2)

# Se añade "ruido"
yn = y + 0.9 * np.random.normal(size=len(x))

# Se ajusta a una curva
popt, pcov = curve_fit(func, x, yn)


ym = func(x, popt[0], popt[1])
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, c='k', label='Función')
ax.scatter(x, yn)
ax.plot(x, ym, c='r', label='Best fit')
ax.legend(loc='upper left')
```
